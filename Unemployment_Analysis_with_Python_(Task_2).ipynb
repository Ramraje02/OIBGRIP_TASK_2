### **Analyzing Data By Barplot**

```python
plt = px.bar(data, x = 'Region', y = ' Estimated Unemployment Rate (%)',color ='Region',title='Unemployment Rate Region-wise By Barplot',animation_frame =' Date',template='plotly')
plt.update_layout(xaxis = {'categoryorder':'total descending'})
plt.show()
```

```python
plt = px.bar(data,x='Region.1',y=' Estimated Unemployment Rate (%)',color='Region',animation_frame=' Date',title='Unemployment Rate Region-wise by Barplot',template='plotly')
plt.update_layout(xaxis={'categoryorder':'total descending'})
plt.show()
```

### **Analyzing Data By Boxplot**

```python
plt = px.box(data,x='Region.1',y=' Estimated Unemployment Rate (%)',color='Region',title='Unemployment rate By Boxplot',animation_frame=' Date',template='plotly')
plt.update_layout(xaxis={'categoryorder':'total descending'})
plt.show()
```

### **Analyzing Data By Scatterplot**

```python
plt = px.scatter(data, x = 'Region.1',y = ' Estimated Unemployment Rate (%)',color='Region',title='Unemployment Rate By Scatterplot',animation_frame=' Date',template='plotly')
plt.update_layout(xaxis={'categoryorder':'total descending'})
plt.show()
```

### **Analyzing Data By Histogram**

```python
plt = px.histogram(data, x = 'Region.1',y = ' Estimated Unemployment Rate (%)',color='Region',title='Unemployment Rate',template='plotly',animation_frame=' Date')
plt.update_layout(xaxis={'categoryorder':'total descending'})
plt.show()
```
